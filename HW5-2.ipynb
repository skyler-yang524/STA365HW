```python
from scipy import stats
import matplotlib.pyplot as plt
import numpy as np
```

### 1. code on normal data: theta prior- normal distribution, tau prior- gamma distribution.

```python
#1. Pymc to get bayesian inference for a sample of normal data where prior for theta is normal distribution and\
#prior for tau is gamma distribuiton. 
n = 100
true_theta, true_tau = 0,1
data_generating_mechanism = \
stats.norm(loc=true_theta, scale=true_tau**(-0.5))

np.random.seed(365) 

x = data_generating_mechanism.rvs(size=n)
plt.hist(x, density=True)
x_grid = np.linspace(-5*true_tau**(-0.5),5*true_tau**(-0.5),1000)
plt.plot(x_grid,data_generating_mechanism.pdf(x_grid));
```

```python
normal_gamma = pymc.Model()

with normal_gamma:

    # Priors for unknown model parameters
    
    theta0 = 0  # prior belief regarding true_theta
    theta_prior_n = 1  # strength of prior belief as units of data
    tau0 = theta_prior_n*true_tau  
    theta = pymc.Normal("theta", mu=theta0, sigma=tau0**(-0.5))
  
    tau_SS_prior = 1  # prior belief regarding true_sum_of_squares
    tau_prior_n = 1  # strength of prior belief as units of data
    
    #gamma distribution not for scale in pymc , we have "rate"
    tau = pymc.Gamma("tau", alpha=tau_prior_n/2, 
                            beta=tau_SS_prior/2)
  
    # Likelihood (sampling distribution) of observations
    x_obs = pymc.Normal("x_obs", mu=theta, tau=tau, observed=x)
    #dependent hirerachy create here. Universal sampling here
```

```python
pymc.model_to_graphviz(normal_gamma)
```

```python
m = 10000
with normal_gamma:
    # draw m posterior samples
    idata_1 = pymc.sample(draws=m, chains=2, tune=100)
    # arviz.InferenceData object
```

```python
theta = idata_1.posterior['theta'].values#.shape
tau = idata_1.posterior['tau'].values#.shape

fig,ax = plt.subplots(1, 4, figsize=(14,3));ax[0].set_title("$\\theta$ Burn-In"); ax[1].set_title("$\\tau$ Burn-In"); ax[2].set_title("$\\theta$ Converged Markov Chain\nStationary Distribution Samples"); ax[3].set_title("$\\tau$ Converged Markov Chain\nStationary Distribution Samples")

# pymc.sample(draws=1000, chains=2, tune=100)
# used 100 samples per chain to "tune" and these were
# automatically discarded so `burn` is likely not needed
burn = 20
demo = 120
C = 2
for c in range(C):
    ax[0].plot(theta[c,:burn], label="$\\theta$ Chain "+str(c))
    ax[1].plot(tau[c,:burn], label="$\\tau$ Chain "+str(c))
    ax[2].plot(np.arange(burn, demo, dtype=int), theta[c,burn:demo], label="$\\theta$ Chain "+str(c))
    ax[3].plot(np.arange(burn, demo, dtype=int), tau[c,burn:demo], label="$\\tau$ Chain "+str(c))
ax[0].legend(); ax[1].legend(); ax[2].legend(); ax[3].legend();  
```

```python
fig,ax = plt.subplots(1, 3, figsize=(16,4))
ax[0].set_title("Marginal distribution of $\\theta$ (without burn-in)\nof (joint posterior) stationary distribution")
ax[1].set_title("Marginal distribution of $\\tau$ (without burn-in)\nof (joint posterior) stationary distribution")
ax[2].set_title("Joint (posterior) stationary distribution\n(without burn-in)")

ax[0].hist(theta[0,burn:])
ax[1].hist(tau[0,burn:])
ax[2].plot(theta[0,burn:], tau[0,burn:], '.', alpha=0.1);
demo = 3*burn
ax[2].plot(theta[0,burn:demo], tau[0,burn:demo]);
```

### 2. code on normal data: theta prior- non-normal distribution, tau prior- non-gamma distribution.

```python
n = 100
true_theta, true_tau = 0,1
data_generating_mechanism = \
stats.norm(loc=true_theta, scale=true_tau**(-0.5))
np.random.seed(365) 

x = data_generating_mechanism.rvs(size=n)
plt.hist(x, density=True)
x_grid = np.linspace(-5*true_tau**(-0.5),5*true_tau**(-0.5),1000)
plt.plot(x_grid,data_generating_mechanism.pdf(x_grid));
```

Instead of using normal prior for theta, we use uniform distribution for theta prior; and also consider to use exponential distribution for tau prior. 

```python
uniform_expon = pymc.Model()

with uniform_expon:

    #non normal theta prior with lower and upper bound of -1,1
    theta = pymc.Uniform("theta", lower=-1, upper=1)  
    #non gamma tau prior, set lambda of exponential distribution as 1.
    tau = pymc.Exponential("tau", lam=1.0)  

    # Likelihood (sampling distribution) of observations
    x_obs = pymc.Normal("x_obs", mu=theta, tau=tau, observed=x)
```

### Q3: Slice Sampling

*First explain how the Markov algorithm of slice sampling as given below works. Then explain the steps by which slice sampling could be used in place of a Metropolis-Hasting step in a Metropolis within Gibbs algorithm where the full conditionals are only known up to a normalizing constant. In your explanation clarify what the curve that we're sampling beneath is, and what the initial value and steps are to create the draw for Gibbs sampling.*

### Answer:

How Markove algorithm of slice sampling works: The first slice_f_at_y function is used to generate a new sample of x based on the given sample x and defined horizontal slice y. It first finds the interval [a,b] where the density function f(x) is greater than y, using grid-based search (x_grid). Then, the new sample x_ is drawn uniformly from the generated interval [a,b]. If the value of density function of new drawn sample x_ is greater than y, then the new sample x_ is accepted and new sample x_ is under f. Otherwise, if new sample x_ is above f and is to the left of x, then left boundary a is moved to new sample x_; if new sample x_ is above f and is to the right of x, then right boundary b is moved to new sample x_. And the function will calls itself to retry with the refined interval, using defined helper function slice_f_at_y_.

Explain the steps by which slice sampling used in place of MH step in a Metropolis with Gibbs: in case of the full conditionals are only known up to a normalizing constant, Metropolis-Hasting could be used which requires a proposal distribution. Instead, slice sampling can be used to directly sample from the conditional without needing a proposal. 
Here's how it works according to the code provided: first identify the target full density f as beta(1.5,3) distribution and plot the desired distribution. Then define slice y is drawn from uniform distribution (y~ Uniform(0,f(x_t))), this y is the horizontal slice under the curve. Then sample the new x_ using the defined function "slice_f_at_y" and finally store the new x_ samples. Plots the histogram of the sampled values 
x_ and as could be seen from the plot, the samples approximate the Beta(1.5,3) distribution. Based on the code provided below, the curve we are sampling beneath is Beta(1.5,3) distribution. The initial value used is x[:2,0] = 0.25. 

```python
def slice_f_at_y(f, x, y, x_grid=np.linspace(0,1,51)):
    
    # find interval of grid points where f(x_grid) > y 
    # then extend the enterval so endpoints f(a)<y and f(b)<y 
    x_grid_delta = x_grid[1]-x_grid[0]
    a,b = x_grid[f(x_grid)>y][[0,-1]]+[-x_grid_delta,x_grid_delta]
    # a,b = x_grid[0,-1]  # make the interval all of x_grid
    
    x_ = a + stats.uniform().rvs()*(b-a)    
    if f(x_)>y:
        return x_,1  # in 1 try if f(x_)>y and "x_ is under f" 
    elif x_ < x:  # or if "x_ was above f on the left side of the interval" 
        x_l,x_r = x_,b
    else:
        x_l,x_r = a,x_  # or if "x_ was above f on the right side of the interval"
    return slice_f_at_y_(f, x, y, x_l, x_r, tot = 2)  # try again with a reduced interval

def slice_f_at_y_(f, x, y, x_l=0, x_r=1, tot=1):
    
    x_ = x_l + stats.uniform().rvs()*(x_r-x_l)
    
    if f(x_)>y:
        return x_,tot
    elif x_ < x:
        x_l = x_
    else:
        x_r = x_
    return slice_f_at_y_(f, x, y, x_l, x_r, tot = tot+1)
```

```python
x_grid = np.linspace(0,1,1000)
f = lambda x: stats.beta(1.5,3).pdf(x)
plt.plot(x_grid, f(x_grid))
    
m = 1000
x = np.zeros([m+1,3])
x[:2,0] = 0.25

plot_trace = 10
for t in range(1,m):
    
    x[t,1] = stats.uniform().rvs()*f(x[t,0])
    if t < plot_trace:
        plt.plot([x[t,0]]*2, [x[t-1,1],x[t,1]], 'k')
    
    x[t+1,0],x[t+1,2] = slice_f_at_y(f, x[t,0], x[t,1])
    if t < plot_trace:
        if t==1:
            plt.plot([x[t,0], x[t+1,0]], [x[t,1]]*2, 'k--', label=str(plot_trace)+ " iterations")
        plt.plot([x[t,0], x[t+1,0]], [x[t,1]]*2, 'k--')
    
plt.hist(x[:,0], density=True, label=str(m)+" iterations\n x values only")
plt.legend();
```
