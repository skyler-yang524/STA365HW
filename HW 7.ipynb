### Q1: derive the following

1. $p(\boldsymbol \beta | \textbf{y}, \textbf{X}, \Sigma=\sigma^2 I)$ for the **linear regression model** likelihood proportional to $\exp \left(-{\frac {1}{2}}(\mathbf {y} -\mathbf{X\boldsymbol \beta})^\top \Sigma^{-1}(\mathbf {y} - \mathbf{X\boldsymbol\beta})\right)$ and a $\mathcal{MVN}(\boldsymbol\beta_0, \boldsymbol\Sigma_\beta)$ for $\boldsymbol\beta$

2. The analytical posterior distribution for $\sigma^2$ the **error variance** of a **linear regression model** with **design matrix** $\mathbf{X}$ assuming $\sigma^2$ has an **inverse-gamma** prior distribution with parameters $\alpha^*$ and $\beta^*$ (unrelated to $\boldsymbol \beta$).

### Answer: 

1: With question given that the likelihhod is proportional to $\exp \left(-{\frac {1}{2}}(\mathbf {y} -\mathbf{X\boldsymbol \beta})^\top \Sigma^{-1}(\mathbf {y} - \mathbf{X\boldsymbol\beta})\right)$ and prior distribution for $\beta$ follows $\mathcal{MVN}(\boldsymbol\beta_0, \boldsymbol\Sigma_\beta)$: 

Compute the Posterior Distribution $( p(\boldsymbol{\beta} | \mathbf{y}, \mathbf{X}, \sigma^2) $)

Using Bayesâ€™ theorem:

$$
p(\boldsymbol{\beta} | \mathbf{y}, \mathbf{X}, \sigma^2) \propto p(\mathbf{y} | \boldsymbol{\beta}, \mathbf{X}, \sigma^2) p(\boldsymbol{\beta}).
$$

the likelihood function is:

$$
p(\mathbf{y} | \boldsymbol{\beta}) \propto \exp\left( -\frac{1}{2\sigma^2} (\mathbf{y} - \mathbf{X} \boldsymbol{\beta})^T (\mathbf{y} - \mathbf{X} \boldsymbol{\beta}) \right).
$$

The prior for $(\boldsymbol{\beta}$) is:

$$
p(\boldsymbol{\beta}) \propto \exp\left( -\frac{1}{2} (\boldsymbol{\beta} - \boldsymbol{\beta}_0)^T \Sigma_{\beta}^{-1} (\boldsymbol{\beta} - \boldsymbol{\beta}_0) \right).
$$

Now, we complete the square in the exponent:

$$
-\frac{1}{2\sigma^2} (\mathbf{y} - \mathbf{X} \boldsymbol{\beta})^T (\mathbf{y} - \mathbf{X} \boldsymbol{\beta}) - \frac{1}{2} (\boldsymbol{\beta} - \boldsymbol{\beta}_0)^T \Sigma_{\beta}^{-1} (\boldsymbol{\beta} - \boldsymbol{\beta}_0).
$$



The likelihood term is: 
$$
-\frac{1}{2\sigma^2} \mathbf{y}^T \mathbf{y} + \frac{1}{\sigma^2} \mathbf{y}^T \mathbf{X} \boldsymbol{\beta} - \frac{1}{2\sigma^2} \boldsymbol{\beta}^T \mathbf{X}^T \mathbf{X} \boldsymbol{\beta}.
$$ 
The prior term is: 

$$
-\frac{1}{2} \boldsymbol{\beta}^T \Sigma_{\beta}^{-1} \boldsymbol{\beta} + \boldsymbol{\beta}^T \Sigma_{\beta}^{-1} \boldsymbol{\beta}_0 - \frac{1}{2} \boldsymbol{\beta}_0^T \Sigma_{\beta}^{-1} \boldsymbol{\beta}_0.
$$

Then we group the terms that only depend on $\beta$: 

$$
-\frac{1}{2} \left[ \boldsymbol{\beta}^T (\mathbf{X}^T \mathbf{X} / \sigma^2 + \Sigma_{\beta}^{-1}) \boldsymbol{\beta} - 2 \boldsymbol{\beta}^T (\mathbf{X}^T \mathbf{y} / \sigma^2 + \Sigma_{\beta}^{-1} \boldsymbol{\beta}_0) \right].
$$


We have the identity: 
$$
-\frac{1}{2} \boldsymbol{\beta}^T A \boldsymbol{\beta} + \boldsymbol{\beta}^T \mathbf{b} = -\frac{1}{2} (\boldsymbol{\beta} - \boldsymbol{\mu}_{\beta | y})^T A (\boldsymbol{\beta} - \boldsymbol{\mu}_{\beta | y}) + \frac{1}{2} \boldsymbol{\mu}_{\beta | y}^T A \boldsymbol{\mu}_{\beta | y}.
$$

where:

$$
\boldsymbol{\mu}_{\beta | y} = A^{-1} \mathbf{b} = \Sigma_{\beta | y} \left( \mathbf{X}^T \mathbf{y} / \sigma^2 + \Sigma_{\beta}^{-1} \boldsymbol{\beta}_0 \right),
$$

$$
\Sigma_{\beta | y} = A^{-1} = \left(\mathbf{X}^T \mathbf{X} / \sigma^2 + \Sigma_{\beta}^{-1} \right)^{-1}.
$$

Thus, we obtain the standard MVN form:

$$
p(\boldsymbol{\beta} | \mathbf{y}) \propto \exp\left( -\frac{1}{2} (\boldsymbol{\beta} - \boldsymbol{\mu}_{\beta | y})^T \Sigma_{\beta | y}^{-1} (\boldsymbol{\beta} - \boldsymbol{\mu}_{\beta | y}) \right).
$$

This shows that the posterior follows:

$$
\boldsymbol{\beta} | \mathbf{y} \sim \mathcal{MVN} \left( \Sigma_{\beta | y} (\mathbf{X}^T \mathbf{y} / \sigma^2 + \Sigma_{\beta}^{-1} \boldsymbol{\beta}_0), \Sigma_{\beta | y} \right).
$$



2: Deriving the Posterior Distribution for $( \sigma^2 $) (Error Variance)
We assume the prior for $( \sigma^2 $) follows an Inverse-Gamma distribution:

$$
p(\sigma^2) = IG(\alpha^*, \beta^*),
$$

which has the density function:
$$
p(\sigma^2) \propto (\sigma^2)^{-\alpha^* - 1} \exp(-\beta^*/\sigma^2).
$$

Given the likelihood, we couldnt ignore $\sigma^2$ term in this case as $\sigma^2$ is the parameter we are interested.

$$
p(\mathbf{y} | \boldsymbol{\beta}, \sigma^2) \propto (\sigma^2)^{-n/2} \exp\left(-\frac{1}{2\sigma^2} (\mathbf{y} - \mathbf{X} \boldsymbol{\beta})^T (\mathbf{y} - \mathbf{X} \boldsymbol{\beta}) \right),
$$

the posterior is:

$$
p(\sigma^2 | \mathbf{y}) \propto (\sigma^2)^{-n/2} \exp\left(-\frac{S}{2\sigma^2}\right) \times (\sigma^2)^{-\alpha^* - 1} \exp(-\beta^*/\sigma^2).
$$

where $( S = (\mathbf{y} - \mathbf{X} \boldsymbol{\beta})^T (\mathbf{y} - \mathbf{X} \boldsymbol{\beta}) $).

Combining terms,

$$
p(\sigma^2 | \mathbf{y}) \propto (\sigma^2)^{-(\alpha^* + n/2) - 1} \exp\left(-\frac{S/2 + \beta^*}{\sigma^2}\right).
$$

This matches the form of an **Inverse-Gamma distribution**:

$$
\sigma^2 | \mathbf{y} \sim IG\left(\alpha^* + \frac{n}{2}, \beta^* + \frac{1}{2} (\mathbf{y} - \mathbf{X} \boldsymbol{\beta})^T (\mathbf{y} - \mathbf{X} \boldsymbol{\beta}) \right).
$$

### Q2 data source
The following dataset is from Kaggle: this is the dataset about Ecommerce Consumer Behavior Analysis Data. I would use consumer's age as x values and purchase amount as y values.

```python
import mlcroissant as mlc
import pandas as pd

# Fetch the Croissant JSON-LD
croissant_dataset = mlc.Dataset('https://www.kaggle.com/datasets/salahuddinahmedshuvo/ecommerce-consumer-behavior-analysis-data/croissant/download')

# Check what record sets are in the dataset
record_sets = croissant_dataset.metadata.record_sets
print(record_sets)

# Fetch the records and put them in a DataFrame
record_set_df = pd.DataFrame(croissant_dataset.records(record_set=record_sets[0].uuid))
record_set_df.head()
```

```python
from sklearn.preprocessing import StandardScaler

#data cleaning step
# Convert byte strings to normal strings and strip spaces
record_set_df = record_set_df.applymap(lambda x: x.decode("utf-8") if isinstance(x, bytes) else x)

# Convert relevant columns to numeric types
record_set_df["Ecommerce_Consumer_Behavior_Analysis_Data.csv/Age"] = pd.to_numeric(
    record_set_df["Ecommerce_Consumer_Behavior_Analysis_Data.csv/Age"], errors="coerce"
)

record_set_df["Ecommerce_Consumer_Behavior_Analysis_Data.csv/Purchase_Amount"] = pd.to_numeric(
    record_set_df["Ecommerce_Consumer_Behavior_Analysis_Data.csv/Purchase_Amount"].str.replace("$", "").str.replace(",", ""), 
    errors="coerce"
)

# Drop missing values (if any)
record_set_df.dropna(inplace=True)

# Rename columns for easier access
record_set_df.rename(columns={
    "Ecommerce_Consumer_Behavior_Analysis_Data.csv/Age": "Age",
    "Ecommerce_Consumer_Behavior_Analysis_Data.csv/Purchase_Amount": "Purchase_Amount"
}, inplace=True)
```

```python

# Select relevant variables: Age (X) and Income/Purchase Amount (y)
X = record_set_df[['Age']].values  # Predictor
y = record_set_df['Purchase_Amount'].values  # Response variable

# Normalize X and y
scaler_x = StandardScaler()
scaler_y = StandardScaler()
X = scaler_x.fit_transform(X)
y = scaler_y.fit_transform(y.reshape(-1, 1)).flatten()  # Standardize y
```

```python
import pymc as pm
import arviz as az
import matplotlib.pyplot as plt

# Define Bayesian Linear Regression Model
with pm.Model() as model:
    # Priors for regression coefficients
    betas = pm.Normal("betas", mu=0, sigma=1, shape=1)  # Single predictor (Age)
    
    # Prior for sigma (Non-inverse-gamma)
    sigma = pm.HalfNormal("sigma", sigma=1)
    
    # Likelihood function
    y_obs = pm.Normal("y", mu=X @ betas, sigma=sigma, observed=y)
    
    # Sampling
    trace = pm.sample(2000, tune=1000, return_inferencedata=True, target_accept=0.95)
```

```python
# Summary statistics of posterior
print(az.summary(trace, hdi_prob=0.95))

# Trace plots to check convergence
az.plot_trace(trace)
plt.show()

# Posterior density plots
az.plot_posterior(trace, var_names=["betas", "sigma"])
plt.show()

# Check autocorrelation
az.plot_autocorr(trace, var_names=["betas"])
plt.show()
```

### Q3: perform robust Bayesian Linear Regression

Let $p(\tau_i)$ be $\require{cancel} \textrm{gamma}\big(\tau_i | \alpha = \frac{\nu}{2}, \overset{\textrm{rate}\xcancel{\textrm{scale}}}{\beta = \frac{\nu}{2}}\big)$ and let $p(y_i|\boldsymbol \beta, \tau,\tau_i)$ be $\mathcal{N}(y_i | \textbf{X} \boldsymbol \beta, \overset{\textrm{precision}}{\tau \times \tau_i})$. 

- Return to your dataset and maniputate it to have some various outliers or find another dataset with some outlier data
- Use an appropriate prior for inference on $v$ if you have enough data to do so
- Use the posterior distributions of the $\tau_i$'s to identify data point "outliers" 
- Use the posterior distributions of the $\sigma_i^{-2} = \tau \times \tau_i$ to create posterior distribuitions of the **influence** (the diagonals of the $H$ "hat" matrix $X^\top (X^\top D X)^{-1} X$ where $D_{ij}=0$ and $D_{ii} = \sigma^2_i$) and compare and contras some example "outlier" versus "non outlier" data points

- Provide inference with Bayesian posterior analysis and report MCMC diagnostics

```python
import numpy as np

## first to add some outliers 
# Load dataset
df = record_set_df.copy()

# Introduce outliers by adding extreme values to Purchase Amount
np.random.seed(365)
outlier_indices = np.random.choice(df.index, size=20, replace=False)  # Select indices to add some outliers
df.loc[outlier_indices, "Purchase_Amount"] *= 5  # Inflate purchase amount as outliers

# Normalize the dataset
scaler_x = StandardScaler()
scaler_y = StandardScaler()
X = scaler_x.fit_transform(df[['Age']])  # Keep DataFrame structure
y = scaler_y.fit_transform(df[['Purchase_Amount']])  # Keep DataFrame structure

# Convert `outlier_indices` from original dataset to new integer positions
outlier_positions = df.index.get_indexer(outlier_indices)  # Get integer indices

# Plot data with outliers
plt.scatter(X, y, label="Data Points")
plt.scatter(X[outlier_positions], y[outlier_positions], color='red', label="Outliers", marker='x', s=100)
plt.xlabel("Age")
plt.ylabel("Normalized Purchase Amount")
plt.legend()
plt.title("Data with Outliers")
plt.show()

```

```python
import pymc as pm
import arviz as az

with pm.Model() as robust_model:
    # Regression coefficients
    betas = pm.Normal("betas", mu=0, sigma=1, shape=1)

    # Hierarchical variance, prior distribution of nu follows gamma distri
    tau = pm.HalfNormal("tau", sigma=1)  # Global variance
    nu = pm.Gamma("nu", alpha=2, beta=0.1)  # Degrees of freedom (controls robustness)
    
    # Individual precision variables as indicated.
    tau_i = pm.Gamma("tau_i", alpha=nu/2, beta=nu/2, shape=len(y))
    
    # Likelihood with individual variances
    y_obs = pm.Normal("y", mu=X @ betas, sigma=pm.math.sqrt(tau * tau_i), observed=y.flatten())

    # Sampling
    trace = pm.sample(2000, tune=1000, return_inferencedata=True, target_accept=0.95)
```

```python
# Extract posterior samples of tau_i
tau_i_samples = trace.posterior["tau_i"].mean(dim=("chain", "draw"))

# Sort data points by their tau_i values (higher tau_i indicates an outlier)
outlier_indices = np.argsort(-tau_i_samples.values)[:20]  # Top 5 suspected outliers

# Scatter plot to highlight suspected outliers
plt.scatter(X, y, label="Data Points")
plt.scatter(X[outlier_indices], y[outlier_indices], color='red', label="Suspected Outliers", marker='x', s=100)
plt.xlabel("Age")
plt.ylabel("Normalized Purchase Amount")
plt.legend()
plt.title("Identified Outliers from Posterior Tau_i")
plt.show()
```

```python
# Extract posterior samples of tau_i and tau
tau_i_samples = trace.posterior["tau_i"].mean(dim=("chain", "draw"))
tau_samples = trace.posterior["tau"].mean(dim=("chain", "draw"))

# Compute posterior mean of sigma_i^2 = tau * tau_i
sigma_sq_i = tau_samples * tau_i_samples
print(sigma_sq_i)
```

```python
import numpy as np

# Design matrix (add intercept)
X_design = np.hstack([np.ones((X.shape[0], 1)), X])  # Add intercept term

# Create diagonal weight matrix using sigma_i^2
D = np.diag(sigma_sq_i.values)

# Compute Hat Matrix
H = X_design @ np.linalg.inv(X_design.T @ D @ X_design) @ X_design.T

# Extract influence scores (diagonal elements of H)
influence_scores = np.diag(H)
```

```python
import seaborn as sns

# Select indices of a few outliers and non-outliers
num_examples = 5
outlier_examples = np.argsort(-influence_scores)[:num_examples]  # Top 5 outliers
non_outlier_examples = np.argsort(influence_scores)[:num_examples]  # 5 least influential points

# Extract posterior samples of tau_i
tau_i_posterior = trace.posterior["tau_i"]

# Plot posterior distributions
plt.figure(figsize=(10, 5))
for idx in outlier_examples:
    sns.kdeplot(tau_i_posterior.sel(tau_i_dim_0=idx).values.flatten(), label=f"Outlier {idx}", linestyle="dashed")

for idx in non_outlier_examples:
    sns.kdeplot(tau_i_posterior.sel(tau_i_dim_0=idx).values.flatten(), label=f"Non-Outlier {idx}")

plt.xlabel("Posterior Tau_i (Variance Scaling)")
plt.ylabel("Density")
plt.legend()
plt.title("Comparison of Posterior Tau_i Distributions: Outliers vs. Non-Outliers")
plt.show()
```

```python
# Summary of posterior distributions
print(az.summary(trace, hdi_prob=0.95))

# Trace plots
az.plot_trace(trace, var_names=["betas", "tau", "nu"])
plt.show()

# Posterior density of tau_i
az.plot_posterior(trace, var_names=["tau_i"])
plt.show()
```
